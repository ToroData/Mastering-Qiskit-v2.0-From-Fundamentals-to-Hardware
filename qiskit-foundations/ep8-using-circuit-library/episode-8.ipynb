## Using the Qiskit Circuit Library

Learning goals:

- Understand what the Qiskit Circuit Library is.
- Learn why and when to use prebuilt gates and circuits.
- Explore parameterization and integration of gates.
- Practice with examples.

## 1. What it is

`qiskit.circuit.library` is a collection of prebuilt gates and circuits.

**Examples:**
- `XGate()`
- `HGate()`
- `RYGate(theta)`
- `CXGate()`
- `SwapGate()`

## 2. Why use it
- Speeds up development.
- Keeps your code modular.
- Reduces implementation errors.

````python
from qiskit.circuit.library import XGate, RYGate
from qiskit import QuantumCircuit
from math import pi

qc = QuantumCircuit(1)
qc.append(XGate(), [0])
qc.append(RYGate(pi/4), [0])
qc.draw('mpl')
````

## 3. Parameterization
- Most gates accept parameters (angles, controls, targets).
- Easy to reuse in different contexts.

````python
from qiskit.circuit.library import RYGate

def rotation_block(qc, theta):
    qc.append(RYGate(theta), [0])
    return qc

# Example usage:
qc = QuantumCircuit(1)
rotation_block(qc, pi/4)
rotation_block(qc, pi/2)
qc.draw('mpl')
````

## 4. Integration
- Append library gates to your existing circuits.
- Combine with custom gates for flexibility.

````python
from qiskit.circuit.library import HGate, CXGate

qc = QuantumCircuit(2)
qc.append(HGate(), [0])
qc.append(CXGate(), [0, 1])
qc.measure_all()
qc.draw('mpl')
````

## 5. Optional execution in Aer

````python
from qiskit_aer import AerSimulator

sim = AerSimulator()

job = sim.run(qc)
result = job.result()
result.get_counts()
````

````python
from qiskit.visualization import plot_histogram

plot_histogram(result.get_counts())
````
